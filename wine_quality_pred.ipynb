```python
selected_col = ['volatile acidity','citric acid','sulphates','alcohol','quality']
```

```python
new_df = df[selected_col]
```

```python
new_df.head(15)
```

```python
x = new_df.drop(columns='quality',axis=1)
y = new_df['quality']
```

```python
y = y.apply(lambda x:1 if x >= 7 else 0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
print(acc)
```

```python
result = model.predict([[0.70,0.00,0.56,9.4]])
```

```python
if result == 0:
    print(result[0],"Wine quality is bad")
else:
    print(result[0],"Wine quality is good")
```
